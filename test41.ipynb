```python
from pathlib import Path
import os
import sys
import shutil
import subprocess

# !!! Replace with your GitHub repo URL
REPO_URL = "https://github.com/misssglory/yolo41.git"

WORKDIR = Path.cwd()
REPO_DIR = WORKDIR / "yolov3_chess_homework_lesson"

if not REPO_DIR.exists():
    subprocess.run(["git", "clone", REPO_URL, str(REPO_DIR)], check=True)
else:
    print(f"Repo already exists: {REPO_DIR}")
    subprocess.run(["git", "-C", str(REPO_DIR), "pull", "--ff-only"], check=False)

os.chdir(REPO_DIR)
sys.path.insert(0, str(REPO_DIR / "src"))

print("Repo dir:", REPO_DIR)
print("Python:", sys.executable)

# Install dependencies into current notebook kernel environment.
# If uv exists, use uv. Otherwise fallback to pip.
if shutil.which("uv"):
    print("Installing with uv...")
    subprocess.run(
        ["uv", "pip", "install", "--python", sys.executable, "-r", "requirements.txt", "-e", "."],
        check=True,
    )
else:
    print("Installing with pip...")
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-r", "requirements.txt", "-e", "."],
        check=True,
    )
```
